```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
import seaborn as sns
```

```python
data = pd.read_csv('titanic.csv')
data.head(10)
```

### Data Understanding 

```python
data.info()
```

### Cleaning data

```python
data.drop(['Name' , 'PassengerId' , 'Ticket' , 'Fare', 'Cabin'], axis=1 , inplace=True)
```

```python
data.isnull().sum()
```

```python
mean_of_age = np.round(data['Age'].mean() , decimals=2)
data['Age'].fillna(mean_of_age , inplace=True)
print(mean_of_age)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

### Processing data

#### *Converting categrical Features*

```python
# exchange categorical data to numbers 
le = preprocessing.LabelEncoder()
lebeling = le.fit_transform(data['Sex'])
# print(lebeling)

data['Sex'] = lebeling
print(data)
```

```python
print(le.classes_)
```
